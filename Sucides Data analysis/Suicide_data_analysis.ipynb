```python
age_5 = df.loc[df.loc[:, 'age'] == '5-14 years', :]
age_15 = df.loc[df.loc[:, 'age'] == '15-24 years', :]
age_25 = df.loc[df.loc[:, 'age'] == '25-34 years', :]
age_35 = df.loc[df.loc[:, 'age'] == '35-54 years', :]
age_55 = df.loc[df.loc[:, 'age'] == '55-74 years', :]
age_75 = df.loc[df.loc[:, 'age'] == '75+ years', :]

sns.lineplot(x='year', y='suicides_no', data=age_5)
sns.lineplot(x='year', y='suicides_no', data=age_15)
sns.lineplot(x='year', y='suicides_no', data=age_25)
sns.lineplot(x='year', y='suicides_no', data=age_35)
sns.lineplot(x='year', y='suicides_no', data=age_55)
sns.lineplot(x='year', y='suicides_no', data=age_75)
```

```python
male_df = df.loc[df.loc[:, 'sex'] == 'male', :]
female_df = df.loc[df.loc[:, 'sex'] == 'female', :]

sns.lineplot(x='year', y='suicides_no', data=male_df)
sns.lineplot(x='year', y='suicides_no', data=female_df)
```

```python
year_wise = df[['country', 'year', 'suicides_no']].groupby(by=['country', 'year'], as_index=False).sum()
```

```python
year_wise.sort_values(by='suicides_no', ascending=False)
```

```python
year_wise[year_wise['country']=='Italy'].sort_values(by='suicides_no', ascending=False)
```

```python
russia = year_wise[year_wise['country']=='Russian Federation']
us = year_wise[year_wise['country']=='United States']
japan = year_wise[year_wise['country']=='Japan']
france = year_wise[year_wise['country']=='France']
uk = year_wise[year_wise['country']=='Ukraine']

plt.plot(russia['year'], russia['suicides_no'], label='russia')
plt.plot(us['year'], us['suicides_no'], label='us')
plt.plot(japan['year'], japan['suicides_no'], label='japan')
plt.plot(france['year'], france['suicides_no'], label='france')
plt.plot(uk['year'], uk['suicides_no'], label='uk')
plt.legend()
```
