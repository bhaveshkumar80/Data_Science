```python
df = pd.read_csv("datasets/minwage.csv")

act_min_wage = pd.DataFrame()

for name, group in df.groupby("State"):
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name})
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name}))
act_min_wage.head()
```

```python
act_min_wage = act_min_wage.replace(0, np.NaN).dropna(axis=1)
act_min_wage.head()
```

```python
def get_min_wage(year, state):
    try:
        return act_min_wage.loc[year][state]
    except:
        return np.NaN
```

```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/diamonds.csv', index_col=0)
df.head()
```

```python
df['cut'].unique()
```

```python
cut_class_dict = {"Fair": 1, "Good": 2, "Very Good": 3,
                  "Premium": 4, "Ideal": 5}
```

```python
df['clarity'].unique()
```

```python
clarity_dict = {"I3": 1, "I2": 2, "I1": 3, "SI2": 4, "SI1": 5,
                "VS2": 6, "VS1": 7, "VVS2": 8, "VVS1": 9, "IF": 10,
               "FL": 11}
```

```python
df['color'].unique()
```

```python
color_dict = {'E': 6, 'I': 2, 'J': 1, 'H': 3, 'F': 5, 'G': 4, 'D': 7}
```

```python
df['cut'] = df['cut'].map(cut_class_dict)
df['clarity'] = df['clarity'].map(clarity_dict)
df['color'] = df['color'].map(color_dict)
df.head()
```

```python
df.tail()
```

```python
from sklearn.utils import shuffle
from sklearn.linear_model import SGDRegressor

df = shuffle(df)
X = df.drop("price", axis=1).values
y = df['price'].values
```

```python
test_size = 200

X_train = X[:-test_size]
y_train = y[:-test_size]

X_test = X[-test_size:]
y_test = y[-test_size:]
```

```python
clf = SGDRegressor(max_iter=1000)
clf.fit(X_train, y_train)

print(clf.score(X_test, y_test))
```

```python
for X, y in list(zip(X_test, y_test))[:10]:
    print(clf.predict([X])[0], y)
```

```python
from sklearn.svm import SVR
clf = SVR()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
for X, y in list(zip(X_test, y_test))[:10]:
    print(f"model predicts {clf.predict([X])[0]}, real value: {y}")
```
